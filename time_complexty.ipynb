```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_vals = np.linspace(1,50,100)
y_xlogx = x_vals * np.log(x_vals)
y_logx = np.log(x_vals)
y_x = x_vals
```

```python
plt.figure(figsize=(8,6))

plt.plot(x_vals, y_xlogx, label='nlog(n)')
plt.plot(x_vals, y_x, label='linear')

plt.xlabel('n')
plt.ylabel('O(n)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x_vals, y_xlogx, label='nlog(n)')
plt.plot(x_vals, y_x, label='linear')
plt.plot(x_vals, y_logx, label='log(n)')
plt.xlabel('n')
plt.ylabel('O(n)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.plot(x_vals, y_xlogx, label='nlog(n)')
plt.plot(x_vals, y_x, label='linear')
plt.plot(x_vals, y_logx, label='log(n)')
plt.plot(x_vals, (x_vals)**2, label='n^2')
plt.xlabel('n')
plt.ylabel('O(n)')
plt.legend()
plt.show()
```
